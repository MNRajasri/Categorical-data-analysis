## <center> 2) Bar chart </center>

#### __Method-1__

- Using matplotlib
  - by implement with dataframe
  - as well as list or array

```python
cdf=visa_df["education_of_employee"].value_counts()
key=cdf.index
value=cdf.values
```

```python
plt.figure(figsize=(5,5))
ax=plt.bar(key,value)
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
plt.bar_label(ax)
plt.show()
```

```python
df=pd.DataFrame({"Education":key,"Count":value})
df
```

```python
plt.figure(figsize=(5,5))
ax=plt.bar("Education","Count",data=df)
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
plt.bar_label(ax)
plt.show()
```

#### __Method-2__

- using seaborn

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(visa_df,x="education_of_employee",order=key)
ax.bar_label(ax.containers[0])
plt.title("Bar chart")
plt.show()
```

#### __Method-3__

- using value count

```python
cdf=visa_df["education_of_employee"].value_counts()
key=cdf.index
value=cdf.values
```

```python
plt.figure(figsize=(5,5))
ax=cdf.plot(kind="bar")
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
ax.bar_label(ax.containers[0])
plt.show()
```

## <center> 3) Relative frequency table </center>

```python
cdf=visa_df["education_of_employee"].value_counts(normalize=True)
cdf
```

## <center> 4) Pie chart </center>

```python
plt.pie(value,labels=key,autopct=f"%0.3f%%",radius=1,startangle=90)
plt.show()
```

```python
plt.pie(value,labels=key,autopct=f"%0.3f%%",explode=[*[0.1]*4]) # explode=[0.1,0.1,0.1,0.1]
plt.title("Pie chart")
plt.show()
```

```python
plt.pie(value,labels=key,autopct=f"%0.3f%%",radius=2,startangle=90) # explode=[0.1,0.1,0.1,0.1]
plt.show()
```
